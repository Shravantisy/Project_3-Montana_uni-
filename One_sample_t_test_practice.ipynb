1. A research study was conducted to examine the differences between older and younger adults on perceived life satisfaction. A pilot study was conducted to examine this hypothesis. Ten older adults (over the age of 70) and ten younger adults (between 20 and 30) were given a life satisfaction test (known to have high reliability and validity). Scores on the measure range from 0 to 60 with high scores indicative of high life satisfaction; low scores indicative of low life satisfaction. The data are presented below.Hint: Use 0.050 as the alpha.

```python
data = {'Older Adults': [45,38,52,48,25,39,51,46,55,46],
        'Younger Adults': [34,22,15,27,37,41,24,19,26,36]}
```

```python
df = pd.DataFrame.from_dict(data)
```

```python
df.hist()
#samp_size = len(data)
#samp_size
```

```python
df_mean = df.mean().mean()
df_mean
```

```python
stats.ttest_ind(df['Older Adults'], df['Younger Adults'])    #Use 0.050 as the alpha i.e we rejecting this Null hypothesis
```

2. 

* The scores of a random sample of 8 students on a physics test are as 
follows: 60, 62, 67, 69, 70, 72, 75, and 78.
Test to see if the sample mean is significantly different using 0.05 as the alpha value. Report the t and p values.

* The researcher realizes that she accidentally recorded the score that should have been 76 as 67. Are these corrected scores significantly different using 0.05 as the alpha? 

```python
#Your code here
#n= 8, 
py_marks = [60, 62, 67, 69, 70, 72, 75, 78]
py_marks_mean1 = np.mean(py_marks)
py_marks_mean1
```

```python
tscore, pvalue = ttest_1samp(a = py_marks, popmean= 1) # tscore 31.56 # we will reject null hypothesis
pvalue# approximetly 0 round(8) 1e-08
```

```python
py_change_marks = [60, 62, 76, 69, 70, 72, 75, 78]
py_marks_mean2 = np.mean(py_change_marks)
py_marks_mean2
```

```python
tscore, pvalue = ttest_1samp(a = py_change_marks, popmean= 1) #tscore 30.25
pvalue
```

The first number for each subject is their performance in the "water" condition. Higher scores reflect better performance. Test to see if alcohol had a significant effect. Use 0.05 as the alpha.
