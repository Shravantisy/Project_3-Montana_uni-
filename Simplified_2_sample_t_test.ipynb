## Manual calculation:

Note: We use a different set of data, but the approach here shows you how scipy calculates the t_statistic and the pvalue.

```python
from math import sqrt
import numpy as np
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from scipy.stats import sem
from scipy.stats import t
from scipy import stats
from scipy.stats import shapiro
import matplotlib
```

```python
# seed the random number generator
seed(1)
# generate two independent samples
data1 = randn(100)
data2 = randn(100)
```

```python
# function for calculating the t-test for two samples
def independent_ttest(data1, data2):
	# calculate means
	mean1, mean2 = mean(data1), mean(data2)
	# calculate standard errors
	se1, se2 = sem(data1), sem(data2)
	# standard error on the difference between the samples
	sed = sqrt(se1**2.0 + se2**2.0)
	# calculate the t statistic
	t_stat = (mean1 - mean2) / sed
	# degrees of freedom
	df = len(data1) + len(data2) - 2
	# calculate the p-value
	pvalue = (1.0 - t.cdf(abs(t_stat), df)) * 2.0
	# return everything
	return t_stat, pvalue
```

```python
# calculate the t test
alpha = 0.05
t_stat, pvalue = independent_ttest(data1, data2)
print('t=%.3f, pvalue=%.3f' % (t_stat,pvalue))

# interpret via p-value
if pvalue > alpha:
	print('Reject null hypothesis that the means are equal.')
else:
	print('Fail to Reject the null hypothesis that the means are equal.')
```
